```python
def entropy(col):
    N=col.shape[0]   # no.of samples
    ent=0.0
    uniques=np.unique(col,return_counts=True)
    for ix in uniques[1]:
        p=ix/N
        ent += (-1.0*p*np.log2(p))
        
    return ent
```

```python
def divide_data(X_data,fkey,fval):
    
    data_left=pd.DataFrame([],columns=X_data.columns)
    data_right=pd.DataFrame([],columns=X_data.columns)
    
    for ix in range(X_data.shape[0]):
        val=X_data[fkey].loc[ix]  
        
        if val > fval:
            data_right.loc[ix]=X_data.loc[ix]
        else:
            data_left.loc[ix]=X_data.loc[ix]
    return data_left,data_right
```

```python
def information_gains(X_data,fkey,fval):
    
    left,right=divide_data(X_data,fkey,fval)
    l=float(left.shape[0])/X_data.shape[0]
    r=float(right.shape[0])/X_data.shape[0]
    
    if left.shape[0]==0 or right.shape[0]==0:
        return -100000
    i_gain= entropy(X_data.Survived) - ((l*entropy(left.Survived))+ (r*entropy(right.Survived)))
    
    return i_gain
```

```python
for fx in X.columns:
    print('The Information Gain of {} is {}'.format(fx,information_gains(df,fx,df[fx].mean())))
```

```python
class DecisionTree:
    
    def __init__(self,depth=0,max_depth=5):
        self.depth=depth
        self.max_depth=max_depth
        self.fkey=None
        self.fval=None
        self.left=None
        self.right=None
        self.target=None
        
    def train(self, data):
        features=['Sex','Age','Parch','SibSp','Pclass','Fare']
        info_gains=[]
        
        for ix in features:
            i_gain=information_gains(data,ix,data[ix].mean())
            info_gains.append(i_gain)
        
        self.fkey=features[np.argmax(info_gains)]
        self.fval=data[self.fkey].mean()
        print("Making Tree using ",self.fkey)
        
        data_left,data_right=divide_data(data,self.fkey,self.fval)
        data_left=data_left.reset_index(drop=True)
        data_right=data_right.reset_index(drop=True)
        
        #Truely a leaf Node
        if data_left.shape[0]==0 or data_right.shape[0]==0:
            if data.Survived.mean()>=0.5:
                self.target="Survive"
            else:
                self.target="Dead"
                
            return 
        
        #Early stopping
        if self.depth>=self.max_depth:
            if data.Survived.mean()>=0.5:
                self.target="Survive"
            else:
                self.target="Dead"
                
            return
        print(self.depth)
        # Recursive call
        self.left=DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.left.train(data_left)
        
        self.right=DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.right.train(data_right)
        
        ## this is for only root nodes
        if data.Survived.mean()>=0.5:
            self.target="Survive"
        else:
            self.target="Dead"
        return
    
    def predict(self,test):
        
        if test[self.fkey]>self.fval:   #go to right
            if self.right is None:
                return self.target
            return self.right.predict(test)
        # go to left
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test)
```
